# Data Exploration: 311 Data - Complaints about Trees
This is a short example of how to start using Pandas (a Python data manipulation package) to explore the 311 Service Request data that was downloaded from NYC Open Data Portal.

Filters: 2016 data and Agency = DPR

### The first step is to do some setup import the relevant python packages:

```python
import pandas as pd
pd.set_option('display.max_columns', 500)

# graphics with matplot lib
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('fivethirtyeight')
```

### Next, read in the data, which was downloaded as a csv file, into a pandas dataframe
* the data file = '311_DPR_2016.csv'
* print out the shape (rows,columns) of the dataframe and 
* show the first 5 rows of data.

```python
sr = pd.read_csv('311_DPR_2016.csv')
print sr.shape
sr.head()
```

Let's take the transpose, so it's easier to see the columns

```python
sr.head().T 
```

### What's in the 'Desciptor' column?
Looking at the first few lines of data, the Descriptor column looks like it might be interesting. Let's take a closer look.

Do a groupby to look at the type of results in that column and sort by number of values, with the most common values listed first.

```python
sr.groupby('Descriptor').count()[['Unique Key']].sort_values('Unique Key', ascending = False)
```

### Pick out all the Fallen Tree complaints
In order to focus in on fallen trees, let's create a new dataframe which only has complaints 'Entire Tree Has Fallen Down'

Note: The .copy() is not really necessary (it basically specifies that any changes to fallenTree will not affect the source dataframe sr), but without it you get annoying warnings from pandas. 

```python
fallenTree = sr[sr['Descriptor']=='Entire Tree Has Fallen Down'].copy()
fallenTree.shape
```

### Date field format
Pandas read in the Date columns as objects, so let's do some reformating.
Format the 'Created Date' field into a datetime format, and then drop the time part.

```python
fallenTree['Created Date'] = pd.to_datetime(fallenTree['Created Date'])
fallenTree['CreatedDate'] = fallenTree['Created Date'].apply(lambda x:x.strftime('%Y-%m-%d'))
```

 ### Plot the frequency counts of the number of complaints per day.

```python
freq = fallenTree.groupby('CreatedDate').count()[['Unique Key']]
freq.plot(legend=False, figsize=(15,8))
plt.ylabel('complaints per day')
plt.title('Frequency of Fallen Tree Complaints via 311')
```

We see a few spikes in the data, where there are more complaints than usual about fallen trees. Let's look at the dates where there are more than 100 complaints per day.

```python
freq[freq['Unique Key']>100]
```
